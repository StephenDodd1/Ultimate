## Group logins data into 15 minute increments

```python
login_groups = []
for time in logins_df['login_time']:
    epoch_time = dt.datetime.strptime(str(time), '%Y-%m-%d %H:%M:%S')
    login_groups.append(int(epoch_time.timestamp() / (15 * 60)))
```

#### The groups are the timestamps epoch time value divided into 15 minute blocks.

```python
login_groups[0:10]
```

```python
logins_df['login_groups'] = login_groups
```

#### Next we will sort our data for a bit better insight. We systematically group our data into weekly data frames.

```python
sorted_logins_df = logins_df.sort_values(by='login_groups')
```

```python
sorted_logins_df.head()
```

```python
group = 0
df_by_week_list = []

df = pd.DataFrame(columns=['login_groups','login_time'])
for row in sorted_logins_df.iterrows():
    current_item_group = int( row[1]['login_groups'] / (24*4*7) )
    if current_item_group == group:
        df.loc[len(df.index)] = [row[1]['login_groups'], row[1]['login_time']] 
    else:
        df.loc[len(df.index)] = [row[1]['login_groups'], row[1]['login_time']]
        if df.shape[0] > 0:
            df_by_week_list.append(df)
        df = pd.DataFrame(columns=['login_groups','login_time'])
        group += 1
```

# Part 2 - Experimental design

```python
# In this scenario, we will assume that there are two cities, which are connected by a 
# toll bridge. The drivers for the company mostly don't travel between the two cities,
# so as to avoid the toll, which consequently, would be paid twice by the driver if 
# crossing the toll bridge, then crossing back. The goal is to encourage drivers to
# cross the bridge, enabling them to have and support more customers, by reimbursing
# the toll fees.
```

####  What would you choose as the key measure of success of this experiment in
#### encouraging driver partners to serve both cities, and why would you choose this metric?


```python
# I'd choose rider wait times as a key measure of success. The purpose of the program 
# is to encourage both inter-city travel among drivers, for instance during peak hours.
# The result should be shorter wait times, particularly on peak hours.
```

#### Describe a practical experiment you would design to compare the effectiveness of the
#### proposed change in relation to the key measure of success. 

```python
# I would test this by running I beta version, offering drivers to pay their toll fees
# for a predetermined period of time, (I'd probably choose three weeks) promising 
# drivers more than usual trips, peak trips in particular, during peak times,when
# traveling to the othe city. Since my data clearly identifies this pattern of early
# peak times in one city and late peak times in the other, it would be to a great 
# advantage.

# In order to test the effectiveness of the program, I'd compare the wait time for both
# a week before and a week after the beta test. In the first week of the beta, I'd 
# begin advertising to drivers, including a two week promotional period in which tolls
# would begin being paid. In the second and third week, the promotion would be on. I
# would compare the data for the week before the beta as a control group, having had no
# intervention at all. The first week of the beta period would be to introduce drivers
# to the idea and prepare them with the details they'll need to know to get started.
# We'd want to make sure drivers understand, and don't hesitate to participate initially
# due to a lack of understanding. It would be ideal to host some sort of Q&A session.

# The two weeks of the beta we would need to note any particular anomalies that could
# corrupt our data. Additionally, we should involve driver check-ins of some type. For
# instance, driver fatigue might lead to drivers taking longer breaks, or something
# else that we didn't account for, which skews our data. Additionally, we'd need to
# ensure that drivers aren't driving for too long so that they are potentially posing
# a hazard. Since we know that the peak hours are complimentary between the two cities,
# we could see drivers attempting to driver for far too many hours, having lost any 
# "slow" periods.

# To test the result I'd plot the entire timeline and run a quadratic regression. 
# The expectation would be to see wait times dip during the two weeks of the test as
# each city is flooded with drivers during peak hours. It was stated that weekends see
# a relatively even spread of activity throughout the day, so we would probably remove
# all weekend days in the event that those days were affecting weekdays, and treat them
# differently. In all likelihood, the wait times would still be reduced on weekends as
# well, just to a smaller degree.

# Assuming that the test was able to reject the null hypothesis, I'd clearly state that
# there is a marked difference in rider wait times, and, assuming that the cost of the
# program isn't too much, the benefits of the program would outweigh it's rejection.

# There are some caveats however. First, we are addressing rider wait times, which
# doesn't account for cancellations, so cancellations would also need to be observed.
# Perhaps prior to the program, more riders were canceling due to wait times, to ride
# with a competitor. This would then skew the previous wait time data in a negative
# direction. In otherwords, fewer cancellations, in the absense of a reduction in 
# wait times, would be indicative of success. As a much greater volume of customers
# with a roughly comparable wait time, while not exactly acheiving the primary goal,
# would still be a huge success, and possibly worth exploring further.
```

## Part 3 - Predictive modeling

```python
df = pd.read_json('../ultimate_data_challenge.json')
```

```python
df.head()
```

```python
new_df = df
last_trip_epoch = []
for date in df['last_trip_date']:
    last_trip_epoch.insert(len(last_trip_epoch), dt.datetime.strptime(str(date), '%Y-%m-%d').timestamp()) 

```

```python
current_time = max(last_trip_epoch)
```

```python
last_trip_days = []
for idx in range(len(last_trip_epoch)):
    current_time_dt = dt.datetime.fromtimestamp(current_time)
    last_trip_days.insert(idx, current_time_dt - dt.datetime.fromtimestamp(last_trip_epoch[idx]))
```

```python
for idx in range(len(last_trip_days)):
    last_trip_days[idx] = str(pd.to_timedelta(last_trip_days[idx], unit='d')).split(' days ')[0]
```

```python
new_df['last_trip_days'] = last_trip_days
```

```python
new_df.info()
```

```python
new_df['last_trip_days'] = new_df['last_trip_days'].astype(int)
new_df['retained'] = new_df['last_trip_days'] < 30
```

```python
'rider retention is ' + str(int(len(new_df[new_df['retained']]) / len(new_df) * 100)) + '%'
```

```python
numeric_cols = new_df.select_dtypes(include=['int32', 'int64', 'float64']).columns
half_cols_length = int(len(numeric_cols) / 2)
print(half_cols_length)
fig, ax = plt.subplots(half_cols_length, 2)
i=0
j=0
for col in numeric_cols:
    ax[j,i].hist(new_df[col], bins = 20)
    ax[j,i].set_title(col)
    ax[j,i].set_yscale('log')
    if j < half_cols_length-1:
        j+=1
    else:
        j=0
        i+=1
fig.set_figheight(12.6)
plt.show()
```

###### Above we can see the average distance is most strongly correlated with last_trip_days, so we will explore that more below. Because we want to predict what will decrease that number, this may not be a very important measure of anything at all.

```python
ax = plt.scatter(new_df['last_trip_days'], new_df['avg_dist'], s=.1)
```

###### Looking at the scatterplot above helps us to see that this correlation is weak at best. We'll remove it. Also we have both average surge and surge percent. We'll drop the one that is less strongly negatively correlated, which is average surge. Lastly, average rating by and of driver are similarly correlated, so we will combine these two columns.

```python
new_df = new_df.drop(columns=['avg_dist', 'avg_surge'])
```

```python
new_df['avg_rating_combined'] = new_df['avg_rating_by_driver'] + new_df['avg_rating_of_driver']
```

```python
new_df = new_df.drop(columns=['avg_rating_of_driver', 'avg_rating_by_driver'])
```

```python
numeric_cols = new_df.select_dtypes(include=['int32', 'int64', 'float64']).columns
sns.heatmap(new_df[numeric_cols].corr())
```

###### My interpretation of the heatmap above is this. The riders who commute using Ultimates service are those most likely to continue using it. Those users, not as frequently, but often, are the same as those with the most trips in the first thirty days. There are almost certainly outliers that confound these results. For instance, a user who owns a car may get the service between selling an old car and buying a new one, or waiting for a vehicle repair. The second class that seems to appear are those that use the service during surge times. That could be weekday service, but is more likely centered around events, which bring more users than is typical.

```python
object_types = new_df.select_dtypes('object').columns
object_types = object_types.drop(['signup_date', 'last_trip_date'])
dummies_df = pd.get_dummies(new_df[object_types])
dummies_df['retained'] = new_df['retained']
print(dummies_df.head())
sns.heatmap(dummies_df.corr())
```

###### Above we can clearly see that the biggest predictors of service usage, this time measure by retained, are iphone users who live in King's Landing. There could be a minor difference in the number of iphone users in Kings Landing that contributes to this correlation. In otherwords, if all iphone users prefer the service, and everyone in King's Landing owns an iphone, then we could safely assume that it's owning an iphone as the predictor, not living in King's Landing in particular. That doesn't appear to be the case. There seems to be only a slight preference for iphone in Kings Landing, probably not enough to cuase this shift. My analysis is that King's Landing is the clear driver her. Users who live in Kings Landing and commute during the week are the most likely to be retained users.

```python
nd_df = dummies_df
```

```python
nd_df['trips_in_first_30_days'] = new_df['trips_in_first_30_days']
nd_df['surge_pct'] = new_df['surge_pct']
nd_df['weekday_pct'] = new_df['weekday_pct']
nd_df['avg_rating_combined'] = new_df['avg_rating_combined']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
rf = RandomForestClassifier(n_estimators=5)
```

```python
nd_df = nd_df.dropna()
X_train, X_test, y_train, y_test = train_test_split(nd_df.drop(columns=['retained']), nd_df['retained']) 
```

```python
X_train
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
rf.score(X_test, y_test)
```

```python
conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```
